```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv('/content/drive/MyDrive/ML/Mall_Customers.csv')
df.head()
```

```python
df.info()
```

```python
df1 = df.iloc[:, 3:5].values
```

```python
import scipy.cluster.hierarchy as sch

plt.figure(figsize = (10, 7))
plt.title('Customer Dendrogram')
dendrogram = sch.dendrogram(sch.linkage(df1, method = 'ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering

model = AgglomerativeClustering(n_clusters = 5,linkage = 'ward')
label_= model.fit_predict(df1)
```

```python
label_
```

```python
plt.figure(figsize = (10,7))
plt.scatter(df1[:,0], df1[:,1], c = model.labels_, cmap = 'rainbow')
```
